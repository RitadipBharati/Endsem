```python
import math
import matplotlib.pyplot as plt
```

```python
T=[1.5,1.8,2.1,2.7,3.0,3.3,3.6,3.9,4.8,5.1,5.4,5.7,6.0,6.6,6.9,7.2,7.8,8.1,8.4,8.7]
S=[0.45,0.42,0.42,0.36,0.33,0.31,0.32,0.31,0.28,0.29,0.24,0.23,0.25,0.21,0.23,0.21,0.23,0.22,0.24,0.23]
```

```python
def Lin_fit(X,Y):
    N=len(X)
    Sx=0
    Sy=0
    Sxx=0
    Sxy=0
    Syy=0
    for i in range(N):
        Sx+=X[i]
        Sy+=Y[i]
        Sxx+=(X[i])**2
        Sxy+=X[i]*Y[i]
        Syy+=(Y[i])**2
    #Ritadip
    m=(N*Sxy-Sx*Sy)/(N*Sxx-(Sx)**2)
    c=Sy/N-m*Sx/N
    YN=[(m*x+c) for x in X]
    r2=((Sxy-Sx*Sy/N)**2)/((Syy-(Sy**2)/N)*(Sxx-(Sx**2)/N))
    if m<0:
        s=-1
    else:
        s=1
    r=s*r2**0.5
    return(X,YN,m,c,r)
```

```python
#Question(i)
LnS=[math.log(s) for s in S]
LnT=[math.log(t) for t in T]
```

```python
X1,Y1,m1,c1,r1=Lin_fit(T,LnS)
print("For sigma=sigma_0*exp(alpha*T): alpha= "+str(m1)+" ,sigma_0= "+str(math.exp(c1))+" ,pearsons r: "+str(r1))
```

```python
plt.plot(X1,Y1,"-.")
plt.scatter(T,LnS)
plt.xlabel("T")
plt.ylabel("ln(sigma)")
plt.grid()
#@Rit1911129
plt.show()
```

```python
#Question(ii)
X2,Y2,m2,c2,r2=Lin_fit(LnT,LnS)
print("For sigma=sigma_0*T^alpha: alpha= "+str(m2)+" ,sigma_0= "+str(math.exp(c2))+" ,pearsons r: "+str(r2))
```

```python
plt.plot(X2,Y2,"-.")
plt.scatter(LnT,LnS)
plt.xlabel("ln(T)")
plt.ylabel("ln(sigma)")
plt.grid()
plt.show()
```
